```python
import pandas as pd
```

```python
grnas = pd.read_excel('gRNAList-extraannot.xlsx').dropna(how = 'all')
grnas.head()
```

```python
import os
from Bio.Seq import reverse_complement


def normalize_grna(inp):
    
    parts = inp.split('.')
    try:
        if len(parts[0]) > len(parts[1]):
            return parts[0]
        else:
            return reverse_complement(parts[1])
    except IndexError:
        return inp
    
```

```python

for _, row in grnas.iterrows():
    
    slug = '%i_%s' % (row['Citation'], row['Name'])
    path = os.path.join('content', 'guides', slug.strip() + '.md')
    
    norm_grna = normalize_grna(row['gRNA'])
    
    with open(path, 'w') as handle:
        handle.write('Template: guide\n')
        handle.write('Title: %i %s\n' % (row['Citation'], row['Name']))
        handle.write('Date: 2019-01-01\n')
        handle.write('grna: %s\n' % norm_grna)
        try:
            handle.write('hxb2_start: %i\n' % row['Start'])
            handle.write('hxb2_stop: %i\n' % row['Stop'])
        except:
            pass
        
        handle.write('target_region: %s\n' % row['Region'])
    
```
